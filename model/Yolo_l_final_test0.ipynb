```python
from ultralytics import YOLO
import os

# 여기 폴더명 입력 (모델 저장 경로와 동일하게 설정)
#custom_folder_name = "m"

# 여기 파일명 입력 (학습된 모델 파일명과 일치하게 설정)
#custom_file_name_prefix = "yolom_imgsz=750_a_epoch"

# 저장된 모델 경로
#model_path = f"/content/drive/MyDrive/Bigproject/model/{custom_folder_name}/{custom_file_name_prefix}_best.pt"
#model_path = '/content/drive/MyDrive/Bigproject/model/yolo_food_exp19/weights/best.pt'
model_path = '/content/drive/MyDrive/Bigproject/model/yolo_food_final2/weights/best.pt'

print(model_path)
#Errno 2] No such file or directory: '/content/drive/MyDrive/Bigproject/model/m/yolom_imgsz=750_a_best.pt'
#yolom_imgsz=750_a_epoch_best.pt
# 데이터셋 설정
data_yaml_path = "/content/drive/MyDrive/Bigproject/data_food/data_food.yaml"

# 모델 로드
model = YOLO(model_path)

# 모델 평가 수행 (validation set 기준)
results = model.val(
    data=data_yaml_path,  # 데이터셋 설정 파일
    imgsz=768,            # 이미지 크기 설정 (학습 시와 동일하게)
    batch=16,             # 배치 크기 설정
    conf=0.25,            # 신뢰도 임계값
    iou=0.45,             # IoU 임계값
    device=0              # GPU 사용
)

```

```python
# 평가 결과 출력
print("\n===== Evaluation Results =====")

# mAP50 및 mAP50-95 출력 (괄호 제거)
print(f"mAP50: {results.box.map50:.3f}")  # mAP@0.5
print(f"mAP50-95: {results.box.map:.3f}")  # mAP@0.5:0.95

# Precision과 Recall 출력
print(f"Precision: {results.box.mp:.3f}")  # Mean Precision
print(f"Recall: {results.box.mr:.3f}")  # Mean Recall
```

```python
# 평가 결과 출력
print("\n===== Evaluation Results =====")
print(f"mAP50: {results.box.map50:.3f}")  # mAP@0.5
print(f"mAP50-95: {results.box.map:.3f}")  # mAP@0.5:0.95

# Precision과 Recall 출력
precision = results.box.mp  # Mean Precision
recall = results.box.mr  # Mean Recall
print(f"Precision: {precision:.3f}")
print(f"Recall: {recall:.3f}")

# F1 Score 계산 및 출력
f1_score = 2 * (precision * recall) / (precision + recall + 1e-8)  # 1e-8 추가하여 zero-division 방지
print(f"F1 Score: {f1_score:.3f}")

# IoU 분석 (평균 IoU 출력)
iou_mean = results.box.iou  # 평균 IoU
print(f"Mean IoU: {iou_mean:.3f}")

# FPS (속도) 측정 및 출력
fps = 1000 / results.speed["inference"]  # ms 단위를 초당 프레임으로 변환
print(f"FPS (Inference Speed): {fps:.2f} frames/sec")
```
