```python
from ultralytics import YOLO
import shutil
import os

# ì—¬ê¸° í´ë”ëª… ì…ë ¥
custom_folder_name = "m"

# ì—¬ê¸° íŒŒì¼ëª… ì…ë ¥
custom_file_name_prefix = "yolol_imgsz=768"

# ì €ì¥ ê²½ë¡œ ì„¤ì •
output_dir = f"/content/drive/MyDrive/Bigproject/model/{custom_folder_name}/"

# í´ë”ê°€ ì¡´ì¬í•˜ì§€ ì•Šìœ¼ë©´ ìƒì„±
os.makedirs(output_dir, exist_ok=True)

# 1. ì‚¬ì „ í•™ìŠµëœ YOLO ëª¨ë¸ ë¡œë“œ
model1 = YOLO('yolo11m.pt')

# 2. í•™ìŠµ ë°ì´í„°ì…‹ ì„¤ì •
data_yaml_path = "/content/drive/MyDrive/Bigproject/data_food/data_food.yaml"

# 3. ëª¨ë¸ í•™ìŠµ ìˆ˜í–‰
results = model1.train(
    data=data_yaml_path,             # ë°ì´í„°ì…‹ ê²½ë¡œ ì„¤ì •
    #imgsz=4,                         # ì´ë¯¸ì§€ í¬ê¸° ì„¤ì •
    imgsz=768,
    batch=16,                         # ë°°ì¹˜ í¬ê¸°
    epochs=200,                         # ì´ í•™ìŠµ ì—í¬í¬ ìˆ˜
    conf=0.25,                        # ì‹ ë¢°ë„ ì„ê³„ê°’
    iou=0.45,                         # IoU ì„ê³„ê°’
    save=True,                        # ìµœì ì˜ ëª¨ë¸ ì €ì¥
    save_period=1,                     # 1 ì—í¬í¬ë§ˆë‹¤ ëª¨ë¸ ì €ì¥
    project="/content/drive/MyDrive/Bigproject/model/",  # ê²°ê³¼ ì €ì¥ ê²½ë¡œ
    name="yolo_food_final_m",              # ì‹¤í—˜ëª… ì§€ì •
    device=0,                          # GPU ì‚¬ìš© (0: ì²« ë²ˆì§¸ GPU)
    half=True                          # FP16ìœ¼ë¡œ í‰ê°€ ì†ë„ í–¥ìƒ (GPU ì‚¬ìš© ì‹œ)
)

# 4. ì €ì¥ëœ ëª¨ë¸ íŒŒì¼ ì´ë¦„ ë³€ê²½ (ì‚¬ìš©ì ì§€ì •)
weights_path = os.path.join(results.save_dir, "weights")

# ì—í¬í¬ë³„ë¡œ ì €ì¥ëœ ëª¨ë¸ íŒŒì¼ ì´ë¦„ ë³€ê²½
for file in os.listdir(weights_path):
    if file.endswith(".pt"):
        epoch_number = file.split('_')[-1].split('.')[0]  # ì˜ˆ: epoch1.pt -> 1 ì¶”ì¶œ
        new_filename = f"{custom_file_name_prefix}_epoch_{epoch_number}.pt"  # ìƒˆë¡œìš´ ì´ë¦„ ì§€ì •
        shutil.move(os.path.join(weights_path, file), os.path.join(output_dir, new_filename))

# ìµœì  ëª¨ë¸ íŒŒì¼ ì´ë¦„ ë³€ê²½
best_model_path = os.path.join(output_dir, "best.pt")
custom_best_model_path = os.path.join(output_dir, f"{custom_file_name_prefix}_best.pt")
if os.path.exists(best_model_path):
    shutil.move(best_model_path, custom_best_model_path)

# ìµœì ì˜ ëª¨ë¸ íŒŒì¼ ê²½ë¡œ ì¶œë ¥
print(f"ëª¨ë¸ì´ '{output_dir}' í´ë”ì— ì €ì¥ë˜ì—ˆìŠµë‹ˆë‹¤.")
print("ëª¨ë¸ í•™ìŠµì´ ì™„ë£Œë˜ì—ˆìŠµë‹ˆë‹¤.")
```
