```python

# 사용자 정의 숫자 설정
label_ind = 22

# 30개의 자연수를 사용자 정의 숫자로 매핑
label_mapping = {i: label_ind for i in range(30)}  # 30개의 자연수만 매핑
print(f"🎯 라벨 매핑: {label_mapping}")

# 변환된 파일 수 카운트
converted_file_count = 0

# labels 폴더가 존재하는지 확인
if os.path.exists(labels_folder):
    # 모든 txt 파일 순회
    for filename in os.listdir(labels_folder):
        if filename.endswith(".txt"):
            file_path = os.path.join(labels_folder, filename)

            # 파일 존재 여부 확인
            if os.path.isfile(file_path):
                # 파일 내용 읽기
                with open(file_path, "r") as file:
                    lines = file.readlines()

                # 각 줄의 첫 번째 숫자 변경
                updated_lines = []
                for line in lines:
                    parts = line.split()
                    if parts:
                        original_label = int(parts[0])
                        # 매핑이 존재하면 변경, 없으면 원래 값 유지
                        new_label = label_mapping.get(original_label, original_label)
                        parts[0] = str(new_label)
                        updated_lines.append(" ".join(parts))

                # 변환된 내용을 파일에 다시 쓰기
                with open(file_path, "w") as file:
                    file.write("\n".join(updated_lines) + "\n")

                # 변환된 파일 수 증가
                converted_file_count += 1

    # 변환된 파일 수 출력
    print(f"✅ 총 {converted_file_count}개의 txt 파일이 성공적으로 변환되었습니다.")
else:
    print(f"❌ 오류: 디렉토리 '{labels_folder}'이(가) 존재하지 않습니다. 경로를 확인하세요.")
```
