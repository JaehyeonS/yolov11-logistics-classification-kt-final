```python
from ultralytics import YOLO
import os

# ì—¬ê¸° í´ë”ëª… ì…ë ¥ (ëª¨ë¸ ì €ì¥ ê²½ë¡œì™€ ë™ì¼í•˜ê²Œ ì„¤ì •)
#custom_folder_name = "m"

# ì—¬ê¸° íŒŒì¼ëª… ì…ë ¥ (í•™ìŠµëœ ëª¨ë¸ íŒŒì¼ëª…ê³¼ ì¼ì¹˜í•˜ê²Œ ì„¤ì •)
#custom_file_name_prefix = "yolom_imgsz=750_a_epoch"

# ì €ì¥ëœ ëª¨ë¸ ê²½ë¡œ
#model_path = f"/content/drive/MyDrive/Bigproject/model/{custom_folder_name}/{custom_file_name_prefix}_best.pt"
#model_path = model_path = '/content/drive/MyDrive/Bigproject/model/yolo_food_exp19/weights/best.pt'
model_path = model_path = '/content/drive/MyDrive/Bigproject/model/yolo_food_final_m2/weights/best.pt'
print(model_path)
#Errno 2] No such file or directory: '/content/drive/MyDrive/Bigproject/model/m/yolom_imgsz=750_a_best.pt'
#yolom_imgsz=750_a_epoch_best.pt
# ë°ì´í„°ì…‹ ì„¤ì •
data_yaml_path = "/content/drive/MyDrive/Bigproject/data_food/data_food.yaml"

# ëª¨ë¸ ë¡œë“œ
model = YOLO(model_path)

# ëª¨ë¸ í‰ê°€ ìˆ˜í–‰ (validation set ê¸°ì¤€)
results = model.val(
    data=data_yaml_path,  # ë°ì´í„°ì…‹ ì„¤ì • íŒŒì¼
    imgsz=768,            # ì´ë¯¸ì§€ í¬ê¸° ì„¤ì • (í•™ìŠµ ì‹œì™€ ë™ì¼í•˜ê²Œ)
    batch=16,             # ë°°ì¹˜ í¬ê¸° ì„¤ì •
    conf=0.25,            # ì‹ ë¢°ë„ ì„ê³„ê°’
    iou=0.45,             # IoU ì„ê³„ê°’
    device=0              # GPU ì‚¬ìš©
)

```

```python
# í‰ê°€ ê²°ê³¼ ì¶œë ¥
print("\n===== Evaluation Results =====")

# mAP50 ë° mAP50-95 ì¶œë ¥ (ê´„í˜¸ ì œê±°)
print(f"mAP50: {results.box.map50:.3f}")  # mAP@0.5
print(f"mAP50-95: {results.box.map:.3f}")  # mAP@0.5:0.95

# Precisionê³¼ Recall ì¶œë ¥
print(f"Precision: {results.box.mp:.3f}")  # Mean Precision
print(f"Recall: {results.box.mr:.3f}")  # Mean Recall
```
