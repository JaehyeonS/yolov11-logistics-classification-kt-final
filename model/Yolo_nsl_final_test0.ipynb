```python
from ultralytics import YOLO
import pandas as pd
import torch
import os

# ‚úÖ Google Drive ÎßàÏö¥Ìä∏ (Ìïú Î≤àÎßå Ïã§Ìñâ ÌïÑÏöî)
from google.colab import drive
drive.mount('/content/drive')

# ‚úÖ Î™®Îç∏ Í≤ΩÎ°ú Î™©Î°ù (Google Drive ÎÇ¥ Í≤ΩÎ°ú ÌôïÏù∏ ÌïÑÏöî)
model_paths = {
    "yolo_food_n": "/content/drive/MyDrive/Bigproject/model/yolo_food_n/weights/best.pt",
    "yolo_food_final_m4": "/content/drive/MyDrive/Bigproject/model/yolo_food_final_m4/weights/best.pt",
    "yolo_food_final_m2": "/content/drive/MyDrive/Bigproject/model/yolo_food_final_m2/weights/best.pt",
    "yolo_food_final2": "/content/drive/MyDrive/Bigproject/model/yolo_food_final2/weights/best.pt"
}

# ‚úÖ Îç∞Ïù¥ÌÑ∞ÏÖã ÏÑ§Ï†ï ÌååÏùº (Google DriveÏóê ÏûàÎäîÏßÄ ÌôïÏù∏)
data_yaml_path = "/content/drive/MyDrive/Bigproject/data_food/data_food.yaml"

# ‚úÖ GPU Í∞ïÏ†ú ÏÇ¨Ïö© (GPUÍ∞Ä ÏóÜÏúºÎ©¥ Ïò§Î•ò Î∞úÏÉù)
if not torch.cuda.is_available():
    raise RuntimeError("üö® CUDA GPUÍ∞Ä Í∞êÏßÄÎêòÏßÄ ÏïäÏïòÏäµÎãàÎã§! ColabÏóêÏÑú 'Îü∞ÌÉÄÏûÑ Ïú†Ìòï Î≥ÄÍ≤Ω'ÏùÑ ÌôïÏù∏ÌïòÏÑ∏Ïöî.")

device = "cuda"
print(f"‚úÖ GPU ÏÇ¨Ïö© Í∞ÄÎä•! Using device: {device}")

# ‚úÖ Í≤∞Í≥º Ï†ÄÏû• Î¶¨Ïä§Ìä∏
results_list = []

# ‚úÖ Í∞Å Î™®Îç∏ ÌèâÍ∞Ä
for model_name, model_path in model_paths.items():
    # Î™®Îç∏ ÌååÏùºÏù¥ Ï°¥Ïû¨ÌïòÎäîÏßÄ ÌôïÏù∏
    if not os.path.exists(model_path):
        print(f"[WARNING] Model file not found: {model_path}. Skipping...")
        continue

    print(f"\nüîç Evaluating Model: {model_name} (Using GPU)")

    # ‚úÖ YOLO Î™®Îç∏ Î°úÎìú
    model = YOLO(model_path)

    # ‚úÖ Î™®Îç∏ ÌèâÍ∞Ä ÏàòÌñâ (Validation Set Í∏∞Ï§Ä)
    results = model.val(
        data=data_yaml_path,
        imgsz=768,
        batch=16,
        conf=0.25,
        iou=0.45,
        device=device  # Î∞òÎìúÏãú GPU ÏÇ¨Ïö©
    )

    # ‚úÖ ÏÑ±Îä• ÌèâÍ∞Ä ÏßÄÌëú Ï†ÄÏû•
    precision = results.box.mp  # Mean Precision
    recall = results.box.mr  # Mean Recall
    f1_score = 2 * (precision * recall) / (precision + recall + 1e-8)  # Zero-division Î∞©ÏßÄ
    iou_mean = (2 * precision * recall) / (precision + recall) if (precision + recall) > 0 else 0
    fps = 1000 / results.speed["inference"]  # FPS (Ï∂îÎ°† ÏÜçÎèÑ)

    # ‚úÖ Í≤∞Í≥º Î¶¨Ïä§Ìä∏Ïóê Ï∂îÍ∞Ä
    results_list.append({
        "Model": model_name,
        "mAP50": round(results.box.map50, 3),
        "mAP50-95": round(results.box.map, 3),
        "Precision": round(precision, 3),
        "Recall": round(recall, 3),
        "F1 Score": round(f1_score, 3),
        "Mean IoU": round(iou_mean, 3),
        "FPS": round(fps, 2)
    })

# # ‚úÖ Í≤∞Í≥ºÎ•º DataFrameÏúºÎ°ú Î≥ÄÌôò Î∞è Ï∂úÎ†•
# df_results = pd.DataFrame(results_list)

# # ‚úÖ Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑ Ï∂úÎ†•
# import ace_tools as tools
# tools.display_dataframe_to_user(name="YOLO Model Performance (GPU)", dataframe=df_results)

# # ‚úÖ ÏµúÏ¢Ö Í≤∞Í≥º Ï∂úÎ†•
# print("\nüìä YOLO Model Performance Comparison (Using GPU):")
# print(df_results)
```

```python

# ‚úÖ Í≤∞Í≥ºÎ•º DataFrameÏúºÎ°ú Î≥ÄÌôò Î∞è Ï∂úÎ†•
df_results = pd.DataFrame(results_list)

# ‚úÖ Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑ Ï∂úÎ†• (ace_tools Ï†úÍ±∞)
from IPython.display import display
display(df_results)

# ‚úÖ CSV ÌååÏùº Ï†ÄÏû• Î∞è Îã§Ïö¥Î°úÎìú
csv_path = "/content/yolo_model_results.csv"
df_results.to_csv(csv_path, index=False)
print(f"\nüìÇ YOLO Î™®Îç∏ ÎπÑÍµê Í≤∞Í≥ºÍ∞Ä CSV ÌååÏùºÎ°ú Ï†ÄÏû•ÎêòÏóàÏäµÎãàÎã§: {csv_path}")

from google.colab import files
files.download(csv_path)

# ‚úÖ ÏµúÏ¢Ö Í≤∞Í≥º Ï∂úÎ†•
print("\nüìä YOLO Model Performance Comparison (Using GPU):")
print(df_results)
```
