```python
from ultralytics import YOLO
import shutil
import os

# Ïó¨Í∏∞ Ìè¥ÎçîÎ™Ö ÏûÖÎ†•
custom_folder_name = "l"

# Ïó¨Í∏∞ ÌååÏùºÎ™Ö ÏûÖÎ†•
custom_file_name_prefix = "yolol_imgsz=768"

# Ï†ÄÏû• Í≤ΩÎ°ú ÏÑ§Ï†ï
output_dir = f"/content/drive/MyDrive/Bigproject/model/{custom_folder_name}/"

# Ìè¥ÎçîÍ∞Ä Ï°¥Ïû¨ÌïòÏßÄ ÏïäÏúºÎ©¥ ÏÉùÏÑ±
os.makedirs(output_dir, exist_ok=True)

# 1. ÏÇ¨Ï†Ñ ÌïôÏäµÎêú YOLO Î™®Îç∏ Î°úÎìú
model1 = YOLO('yolo11l.pt')

# 2. ÌïôÏäµ Îç∞Ïù¥ÌÑ∞ÏÖã ÏÑ§Ï†ï
data_yaml_path = "/content/drive/MyDrive/Bigproject/data_food/data_food.yaml"

# 3. Î™®Îç∏ ÌïôÏäµ ÏàòÌñâ
results = model1.train(
    data=data_yaml_path,             # Îç∞Ïù¥ÌÑ∞ÏÖã Í≤ΩÎ°ú ÏÑ§Ï†ï
    #imgsz=4,                         # Ïù¥ÎØ∏ÏßÄ ÌÅ¨Í∏∞ ÏÑ§Ï†ï
    imgsz=768,
    batch=16,                         # Î∞∞Ïπò ÌÅ¨Í∏∞
    epochs=200,                         # Ï¥ù ÌïôÏäµ ÏóêÌè¨ÌÅ¨ Ïàò
    conf=0.25,                        # Ïã†Î¢∞ÎèÑ ÏûÑÍ≥ÑÍ∞í
    iou=0.45,                         # IoU ÏûÑÍ≥ÑÍ∞í
    save=True,                        # ÏµúÏ†ÅÏùò Î™®Îç∏ Ï†ÄÏû•
    save_period=1,                     # 1 ÏóêÌè¨ÌÅ¨ÎßàÎã§ Î™®Îç∏ Ï†ÄÏû•
    project="/content/drive/MyDrive/Bigproject/model/",  # Í≤∞Í≥º Ï†ÄÏû• Í≤ΩÎ°ú
    name="yolo_food_final",              # Ïã§ÌóòÎ™Ö ÏßÄÏ†ï
    device=0,                          # GPU ÏÇ¨Ïö© (0: Ï≤´ Î≤àÏß∏ GPU)
    half=True                          # FP16ÏúºÎ°ú ÌèâÍ∞Ä ÏÜçÎèÑ Ìñ•ÏÉÅ (GPU ÏÇ¨Ïö© Ïãú)
)

# 4. Ï†ÄÏû•Îêú Î™®Îç∏ ÌååÏùº Ïù¥Î¶Ñ Î≥ÄÍ≤Ω (ÏÇ¨Ïö©Ïûê ÏßÄÏ†ï)
weights_path = os.path.join(results.save_dir, "weights")

# ÏóêÌè¨ÌÅ¨Î≥ÑÎ°ú Ï†ÄÏû•Îêú Î™®Îç∏ ÌååÏùº Ïù¥Î¶Ñ Î≥ÄÍ≤Ω
for file in os.listdir(weights_path):
    if file.endswith(".pt"):
        epoch_number = file.split('_')[-1].split('.')[0]  # Ïòà: epoch1.pt -> 1 Ï∂îÏ∂ú
        new_filename = f"{custom_file_name_prefix}_epoch_{epoch_number}.pt"  # ÏÉàÎ°úÏö¥ Ïù¥Î¶Ñ ÏßÄÏ†ï
        shutil.move(os.path.join(weights_path, file), os.path.join(output_dir, new_filename))

# ÏµúÏ†Å Î™®Îç∏ ÌååÏùº Ïù¥Î¶Ñ Î≥ÄÍ≤Ω
best_model_path = os.path.join(output_dir, "best.pt")
custom_best_model_path = os.path.join(output_dir, f"{custom_file_name_prefix}_best.pt")
if os.path.exists(best_model_path):
    shutil.move(best_model_path, custom_best_model_path)

# ÏµúÏ†ÅÏùò Î™®Îç∏ ÌååÏùº Í≤ΩÎ°ú Ï∂úÎ†•
print(f"Î™®Îç∏Ïù¥ '{output_dir}' Ìè¥ÎçîÏóê Ï†ÄÏû•ÎêòÏóàÏäµÎãàÎã§.")
print("Î™®Îç∏ ÌïôÏäµÏù¥ ÏôÑÎ£åÎêòÏóàÏäµÎãàÎã§.")
```

```python
from ultralytics import YOLO
import os

# Ïó¨Í∏∞ Ìè¥ÎçîÎ™Ö ÏûÖÎ†• (Î™®Îç∏ Ï†ÄÏû• Í≤ΩÎ°úÏôÄ ÎèôÏùºÌïòÍ≤å ÏÑ§Ï†ï)
#custom_folder_name = "m"

# Ïó¨Í∏∞ ÌååÏùºÎ™Ö ÏûÖÎ†• (ÌïôÏäµÎêú Î™®Îç∏ ÌååÏùºÎ™ÖÍ≥º ÏùºÏπòÌïòÍ≤å ÏÑ§Ï†ï)
#custom_file_name_prefix = "yolom_imgsz=750_a_epoch"

# Ï†ÄÏû•Îêú Î™®Îç∏ Í≤ΩÎ°ú
model_path = f"/content/drive/MyDrive/Bigproject/model/{custom_folder_name}/{custom_file_name_prefix}_best.pt"
print(model_path)
#Errno 2] No such file or directory: '/content/drive/MyDrive/Bigproject/model/m/yolom_imgsz=750_a_best.pt'
#yolom_imgsz=750_a_epoch_best.pt
# Îç∞Ïù¥ÌÑ∞ÏÖã ÏÑ§Ï†ï
data_yaml_path = "/content/drive/MyDrive/Bigproject/data_food/data_food.yaml"

# Î™®Îç∏ Î°úÎìú
model = YOLO(model_path)

# Î™®Îç∏ ÌèâÍ∞Ä ÏàòÌñâ (validation set Í∏∞Ï§Ä)
results = model.val(
    data=data_yaml_path,  # Îç∞Ïù¥ÌÑ∞ÏÖã ÏÑ§Ï†ï ÌååÏùº
    imgsz=768,            # Ïù¥ÎØ∏ÏßÄ ÌÅ¨Í∏∞ ÏÑ§Ï†ï (ÌïôÏäµ ÏãúÏôÄ ÎèôÏùºÌïòÍ≤å)
    batch=16,             # Î∞∞Ïπò ÌÅ¨Í∏∞ ÏÑ§Ï†ï
    conf=0.25,            # Ïã†Î¢∞ÎèÑ ÏûÑÍ≥ÑÍ∞í
    iou=0.45,             # IoU ÏûÑÍ≥ÑÍ∞í
    device=0              # GPU ÏÇ¨Ïö©
)

```

```python
# ÌèâÍ∞Ä Í≤∞Í≥º Ï∂úÎ†•
print("\n===== Evaluation Results =====")

# mAP50 Î∞è mAP50-95 Ï∂úÎ†• (Í¥ÑÌò∏ Ï†úÍ±∞)
print(f"mAP50: {results.box.map50:.3f}")  # mAP@0.5
print(f"mAP50-95: {results.box.map:.3f}")  # mAP@0.5:0.95

# PrecisionÍ≥º Recall Ï∂úÎ†•
print(f"Precision: {results.box.mp:.3f}")  # Mean Precision
print(f"Recall: {results.box.mr:.3f}")  # Mean Recall
```

```python
# ÌÅ¥ÎûòÏä§Î≥Ñ ÏÉÅÏÑ∏ ÌèâÍ∞Ä Í≤∞Í≥º Ï∂úÎ†•
print("\n===== Per-Class Evaluation =====")
for cls, ap in enumerate(results.box.maps):
    print(f"Class: {model1.names[cls]}, AP50: {ap:.3f}")

# Í≤∞Í≥º Ï†ÄÏû• ÎîîÎ†âÌÜ†Î¶¨ ÏÉùÏÑ±
output_results_path = f"/content/drive/MyDrive/Bigproject/model/{custom_folder_name}/results/"
os.makedirs(output_results_path, exist_ok=True)

```
