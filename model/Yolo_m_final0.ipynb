```python
from ultralytics import YOLO
import shutil
import os

# 여기 폴더명 입력
custom_folder_name = "m"

# 여기 파일명 입력
custom_file_name_prefix = "yolol_imgsz=768"

# 저장 경로 설정
output_dir = f"/content/drive/MyDrive/Bigproject/model/{custom_folder_name}/"

# 폴더가 존재하지 않으면 생성
os.makedirs(output_dir, exist_ok=True)

# 1. 사전 학습된 YOLO 모델 로드
model1 = YOLO('yolo11m.pt')

# 2. 학습 데이터셋 설정
data_yaml_path = "/content/drive/MyDrive/Bigproject/data_food/data_food.yaml"

# 3. 모델 학습 수행
results = model1.train(
    data=data_yaml_path,             # 데이터셋 경로 설정
    #imgsz=4,                         # 이미지 크기 설정
    imgsz=768,
    batch=16,                         # 배치 크기
    epochs=200,                         # 총 학습 에포크 수
    conf=0.25,                        # 신뢰도 임계값
    iou=0.45,                         # IoU 임계값
    save=True,                        # 최적의 모델 저장
    save_period=1,                     # 1 에포크마다 모델 저장
    project="/content/drive/MyDrive/Bigproject/model/",  # 결과 저장 경로
    name="yolo_food_final_m",              # 실험명 지정
    device=0,                          # GPU 사용 (0: 첫 번째 GPU)
    half=True                          # FP16으로 평가 속도 향상 (GPU 사용 시)
)

# 4. 저장된 모델 파일 이름 변경 (사용자 지정)
weights_path = os.path.join(results.save_dir, "weights")

# 에포크별로 저장된 모델 파일 이름 변경
for file in os.listdir(weights_path):
    if file.endswith(".pt"):
        epoch_number = file.split('_')[-1].split('.')[0]  # 예: epoch1.pt -> 1 추출
        new_filename = f"{custom_file_name_prefix}_epoch_{epoch_number}.pt"  # 새로운 이름 지정
        shutil.move(os.path.join(weights_path, file), os.path.join(output_dir, new_filename))

# 최적 모델 파일 이름 변경
best_model_path = os.path.join(output_dir, "best.pt")
custom_best_model_path = os.path.join(output_dir, f"{custom_file_name_prefix}_best.pt")
if os.path.exists(best_model_path):
    shutil.move(best_model_path, custom_best_model_path)

# 최적의 모델 파일 경로 출력
print(f"모델이 '{output_dir}' 폴더에 저장되었습니다.")
print("모델 학습이 완료되었습니다.")
```
