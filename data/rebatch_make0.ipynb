```python
import os
import shutil
from sklearn.model_selection import train_test_split

# ì›ë³¸ í´ë” ê²½ë¡œ
source_base_path = '/content/drive/MyDrive/Bigproject/data_make'

# ëŒ€ìƒ í´ë” ê²½ë¡œ
target_base_path = '/content/drive/MyDrive/Bigproject/data_food'
target_label_paths = {
    'train': os.path.join(target_base_path, 'labels/train'),
    'val': os.path.join(target_base_path, 'labels/val'),
    'test': os.path.join(target_base_path, 'labels/test')
}
target_image_paths = {
    'train': os.path.join(target_base_path, 'images/train'),
    'val': os.path.join(target_base_path, 'images/val'),
    'test': os.path.join(target_base_path, 'images/test')
}

# ë°ì´í„° ë¹„ìœ¨ ì„¤ì •
train_ratio = 0.6
val_ratio = 0.2
test_ratio = 0.2

# ì²˜ë¦¬í•  í´ë” ëª©ë¡
folders = ['poka0', 'poka', 'hcok', 'coka', 'apple']

# ëŒ€ìƒ í´ë” ìƒì„±
for path_dict in [target_label_paths, target_image_paths]:
    for path in path_dict.values():
        os.makedirs(path, exist_ok=True)

# ë°ì´í„° ë³µì‚¬ í•¨ìˆ˜
def copy_files(files, source_folder, target_folder):
    for file in files:
        source_path = os.path.join(source_folder, file)
        target_path = os.path.join(target_folder, file)
        shutil.copy(source_path, target_path)

# ê° í´ë” ì²˜ë¦¬
for folder in folders:
    label_folder = os.path.join(source_base_path, folder, 'label')
    image_folder = os.path.join(source_base_path, folder, 'image')

    # ë¼ë²¨ ë° ì´ë¯¸ì§€ íŒŒì¼ ëª©ë¡ ê°€ì ¸ì˜¤ê¸°
    label_files = sorted([f for f in os.listdir(label_folder) if f.endswith('.txt')])
    image_files = sorted([f for f in os.listdir(image_folder) if f.endswith(('.jpg', '.png'))])

    # íŒŒì¼ ìˆ˜ê°€ ê°™ì§€ ì•Šìœ¼ë©´ ê²½ê³ 
    if len(label_files) != len(image_files):
        print(f"âš ï¸ {folder}: ë¼ë²¨({len(label_files)})ê³¼ ì´ë¯¸ì§€({len(image_files)}) íŒŒì¼ ìˆ˜ê°€ ì¼ì¹˜í•˜ì§€ ì•ŠìŠµë‹ˆë‹¤.")
        continue

    # train, val, testë¡œ ë¶„í• 
    train_labels, temp_labels, train_images, temp_images = train_test_split(
        label_files, image_files, test_size=(1 - train_ratio), random_state=42)
    val_labels, test_labels, val_images, test_images = train_test_split(
        temp_labels, temp_images, test_size=(test_ratio / (val_ratio + test_ratio)), random_state=42)

    # íŒŒì¼ ë³µì‚¬
    copy_files(train_labels, label_folder, target_label_paths['train'])
    copy_files(train_images, image_folder, target_image_paths['train'])

    copy_files(val_labels, label_folder, target_label_paths['val'])
    copy_files(val_images, image_folder, target_image_paths['val'])

    copy_files(test_labels, label_folder, target_label_paths['test'])
    copy_files(test_images, image_folder, target_image_paths['test'])

    print(f"âœ… {folder}: íŒŒì¼ì´ ì„±ê³µì ìœ¼ë¡œ ë¶„í•  ë° ë³µì‚¬ë˜ì—ˆìŠµë‹ˆë‹¤.")

print("ğŸ‰ ëª¨ë“  ë°ì´í„°ê°€ ì„±ê³µì ìœ¼ë¡œ ì²˜ë¦¬ë˜ì—ˆìŠµë‹ˆë‹¤!")
```
